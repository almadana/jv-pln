```python
from scipy.io import loadmat, savematemat, wavfile
from datasets import load_dataset
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import torch 
```

```python
import matplotlib.pyplot as plt
```

```python
processor = WhisperProcessor.from_pretrained("openai/whisper-medium.en")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-medium.en")

```

```python
samplerate, data = wavfile.read("../aud_data/trials/WAV/trial02.wav")
```

```python
input_features = processor(data,sampling_rate=samplerate, return_tensors="pt").input_features
```

```python
plt.plot(data)
```

```python
plt.plot(input_features[0,1,:])
```

```python
predicted_ids = model.generate(inputs=input_features)
```

```python
transcription = processor.batch_decode(predicted_ids)[0]
```
